```python
Q3. Ans

# Multiple inheritance

Actually, Multiple inheritance is a type of inheritance in which  a child class inherits from more than one parent class.

#Example

class father:
    
    def rahul_father(self):
        
        print("My name is Rohan and My son name is Rahul.")
        
class mother:
    
    def rahul_mother(self):
        
        print("My name is Ravina and My son name is Rahul.")
        
class child(father , mother):
    
    def rahul(self):
        
        print("My name is Rahul and My father's Name is Rohan and My Mother's Name is Ravina.")
        


obj_child=child()

obj_child.rahul_father()
obj_child.rahul_mother()
obj_child.rahul()
```

```python
Q4. Ans

Actually, Getter and Setter are methods in python in which getter method that allows us to access an attribute
or private or protected attributes in a given class, while setter method that allows us to set or mutate the value
of an attributes in given class



class check_getter_and_setter:                 #creating a class
    
    def __init__(self,book_name,price):        #assigning values
        
        self.book_name=book_name
        self.__price=price
        
    @property
    def price_access(self):                   #getter i.e allow to give access a private attribute
        
        return self.__price
    
    @price_access.setter                     #setter i.e allow to set or mutate the value
    def price_update(self,new_price):
        
        self.__price=new_price
        
        return self.__price
        
        
obj=check_getter_and_setter("Let us Python",280)      #creating instance or object

print(obj.book_name)

obj.price_access         # it will give price 280

obj.price_update=265

obj.price_access         # it will give price 265
```

```python
Q5. Ans

Actually, Method overriding is a feature of object-oriented programming languages where the subclass or 
child class can provide the program with specific characteristics or a specific implementation
process of data provided that are already defined in the parent class or superclass.
In simple word,Method overriding in Python is when you have two methods with the same name that each 
perform different tasks.



class test_overriding_parent:
    
    def class_details(self):
        
        print("I am in Parent Class")
        
class test_overriding_child(test_overriding_parent):
    
    def class_details(self):
        
        print("I am in Child Class")
        

obj_overriding=test_overriding_child()

obj_overriding.class_details()

```
