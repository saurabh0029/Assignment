```python
Seaborn is a powerful library for data visualization in Python, which provides a wide range of functions to create 
various types of plots.

# Name any five plots that we can plot using the Seaborn library

1.Scatter Plot - A scatter plot is used to visualize the relationship between two continuous variables. It is useful for
                identifying patterns and  trends, as well as outliers and clusters. Seaborn's scatter plot function can 
                also be used to color-code the points based on a  categorical variable.

2.Line Plot - A line plot is used to visualize the trends and changes in a continuous variable over time or another 
                continuous variable. It is useful for identifying patterns and trends over time, as well as for comparing
                different groups or categories. Seaborn's line plot function can also be used to plot multiple lines on 
                the same plot.

3.Bar Plot - A bar plot is used to visualize the distribution of a categorical variable or the relationship between a 
            categorical variable and a  continuous variable. It is useful for comparing the values of different categories,
            as well as for visualizing the mean or median values of a continuous variable for each category. Seaborn's bar 
            plot function can also be used to plot horizontal bar plots.

4.Heatmap - A heatmap is used to visualize the relationship between two variables in a matrix or dataset, where the values
            are represented as colors. It is useful for identifying patterns and trends in large datasets or matrices, 
            as well as for visualizing the relationships between  different variables or groups. Seaborn's heatmap 
            function can also be used to annotate the cells of the heatmap with the corresponding  values.

5.Box Plot - A box plot is used to visualize the distribution of a continuous variable or the relationship between a
            continuous variable and a  categorical variable. It is useful for identifying the median, quartiles, and 
            outliers of the data, as well as for comparing the  distributions of different categories. Seaborn's box 
            plot function can also be used to plot violin plots, which show the shape of the distribution in addition
            to the quartiles and outliers.
            
            
```

```python
import seaborn as sns  #importing seaborn library

fmri=sns.load_dataset("fmri")  # loading dataset

sns.lineplot(data=fmri , x="timepoint" , y="signal" , hue="event" , style="region") # Creatong lineplot
```
