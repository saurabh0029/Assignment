Q5.Ans

```python


Actually, Dataframe.size  returns the total number of elements in the dataframe, which is equivalent to the number of rows multiplied by
the number of columns. This means that if a dataframe has n rows and m columns, the size of the dataframe will be n x m. The size attribute
returns a single integer value. Whereas dataframe.shape returns a tuple that contains the number of rows and columns in the dataframe. The 
first element of the tuple is the number of rows, and the second element is the number of columns. So, if a dataframe has n rows and m columns,
the shape of the dataframe will be (n, m). The shape attribute returns a tuple of two integers.

```

Q9.Ans

```python
data6={'Values':[2,4,6,7,3,8,9,11,23]} # Given Data

df7=pd.DataFrame(data6)  # Creating DataFrame

def find_mean_median_dtd(df): # Creating Function
    
    mean=df.mean()   # Calculate mean
    
    median=df.median() #Calculate median
    
    std=df.std()  # Calculate Standard Deviatio
    
    return("Mean = {} Median = {} Standard Deviation = {}").format(mean,median,std)

find_mean_median_dtd(df7) # function Calling
```

Q10.Ans

```python
data7={'Sales':[2000,2200,2500,2800,3500,4500,5200,6000,4800,7421,9856,7700,5500,4576],'Date':['1-03-2023','2-03-2023','3-03-2023','4-03-2023',
                '5-03-2023','6-03-2023','7-03-2023','8-03-2023','9-03-2023','10-03-2023','11-03-2023','12-03-2023','13-03-2023','14-03-2023']}

                 #Given Data

df8=pd.DataFrame(data7)  # Creating DataFrame

df8['Date']=pd.to_datetime(df8['Date'],dayfirst=True) # Converting int Data datatype

def Moving_Average(df):  # Creating Function
    
    df.sort_values('Date',inplace=True) # Sorting according to Date
    
    df['Moving Average']=df['Sales'].rolling(window=7,min_periods=1).mean()
    
    return(df)

Moving_Average(df8) # Function Calling
```

```python
Data8={'Values':[1,11,2,3,33,4,5,66,7,67,58,3,4,2,4,5,8,1,2,8,0,4,3,7,44,99,6,7,2,3,4,6,3,778,34,64,57,345,39,40]}

df9=pd.DataFrame(Data8)

df9['Date']= pd.date_range(start ='2023-01-01',end='2023-02-9')

def specific_data(df): # creating Function
    
    start_date = '2023-01-01'
    
    end_date = '2023-01-31'
    
    new_df2 = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    
    return (df[new_df2])

specific_data(df9)
```
