```python
Q5. Ans

use_for() is used in Flask for URL Building.


# code to demonstrate the working of the url_for() function

from flask import *  
   
app = Flask(__name__)  
  
@app.route('/pwskills')  
def pwskills():  
    return '<h1>Welcome to PW Skills <h1>'  
  
@app.route('/iNeuron')  
def iNeuron():  
    return '<h1>Welcome to iNeuron <h1>'  
  


@app.route('/select/<name>')  
def user(name):  
    if name == 'pwskills':  
        return redirect(url_for('pwskills'))  
    if name == 'iNeuron':  
        return redirect(url_for('iNeuron'))  
   
if __name__ =='__main__':  
    app.run(host="0.0.0.0") 
```
