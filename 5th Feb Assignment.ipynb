```python
Q5. Ans

Inheritance is the ability to ‘inherit’ features or attributes from already written classes into newer classes we make.
These features and attributes are defined data structures and the functions we can perform with them.
It promotes code reusability, which is considered one of the best industrial coding practices as 
it makes the codebase modular.

Parent class is the class being inherited from, also called base class.
Child class is the class that inherits from another class, also called derived class.

Types of Inheritance are:-
Single inheritance
Multiple inheritance
Multilevel inheritance
Hierarchical inheritance
Hybrid inheritance

# Single inheritance

When a child class inherits only a single parent class.
#Example

class Parent:
     def func1(self):
          print("I am in function 1")
class Child(Parent):
     def func2(self):
          print("I am in function 2 ")
ob = Child()
ob.func1()
ob.func2()

# Multiple inheritance

When a child class inherits from more than one parent class.

#Example

class Parent:
   def func1(self):
        print("I am in function 1")
class Parent2:
   def func2(self):
        print("I am in function 2")
class Child(Parent , Parent2):
    def func3(self):
        print("I am in function 3")
 
ob = Child()
ob.func1()
ob.func2()
ob.func3()


# Multilevel inheritance

When a child class becomes a parent class for another child class.

#Example

class Parent:
      def func1(self):
          print("I am in function 1")
class Child(Parent):
      def func2(self):
          print("I am in function  2")
class Child2(Child):
      def func3("I am in function  3")
ob = Child2()
ob.func1()
ob.func2()
ob.func3()

# Hierarchical inheritance

Hierarchical inheritance involves multiple inheritance from the same base or parent class.

#Example

class Parent:
      def func1(self):
          print("I am in function  1")
class Child(Parent):
      def func2(self):
          print("I am in function  2")
class Child2(Parent):
      def func3(self):
          print("I am in function 3")
 
ob = Child()
ob1 = Child2()
ob.func1()
ob.func2()

# Hybrid inheritance

Hybrid inheritance involves multiple inheritance taking place in a single program.

#Example

class Parent:
     def func1(self):
         print("I am in function  1")
 
class Child(Parent):
     def func2(self):
         print("I am in function  2")
 
class Child1(Parent):
     def func3(self):
         print(" I am in function  3"):
 
class Child3(Parent , Child1):
     def func4(self):
         print(" I am in function  4")
 
ob = Child3()
ob.func1()
```
