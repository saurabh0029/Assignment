```python
# What are glyphs in Bokeh?

In Bokeh, glyphs are visual shapes used to represent data points in a plot, such as circles, squares, triangles, lines,
and more. They are used to create scatter plots, line plots, and other types of visualizations.


# How can you add them to a Bokeh plot?

To add glyphs to a Bokeh plot, we need to use the figure() function from the bokeh.plotting module. We can then add glyphs
to the plot by calling a glyph method on the figure object, such as circle(), square(), or line(). These methods take 
various arguments, such as x and y coordinates, size, color, and more, to customize the appearance of the glyph.




For Example

from bokeh.plotting import figure, output_file, show
import numpy as np

# generate some random data
x = np.random.normal(size=75)
y = np.random.normal(size=75)

# create a Bokeh plot
output_file("scatter.html")
fig = figure(title="Scatter Plot", x_axis_label="X", y_axis_label="Y")

# add circles as glyphs
fig.circle(x, y, size=10, color="blue") 

# show the plot
show(fig)
```

```python
#What is a Bokeh server?

A Bokeh server is a Python process that allows us to create and serve interactive Bokeh plots to a web browser.
With the Bokeh server,  we can create complex interactive plots that can be updated in real time based on user 
interactions or external data sources.

#how can you use it to create interactive plots that can be updated in real time?


To use the Bokeh server, we need to create a Python script that defines the layout and behavior of our plot. This script
will typically include  a curdoc() function call that returns a Bokeh document object, which contains all the Bokeh models
and widgets that make up our plot.  We can then use the bokeh.server.server.Server class to start the Bokeh server, 
passing in the Bokeh document as the argument.

For example , how to create a simple Bokeh server app that displays a scatter plot with circles that can be resized 
by a slider widget:
    
    from bokeh.io import curdoc
    from bokeh.layouts import column
    from bokeh.models import ColumnDataSource, Slider
    from bokeh.plotting import figure

    # create a Bokeh data source
    source = ColumnDataSource(data=dict(x=[1, 2, 3], y=[2, 4, 6], size=[10, 20, 30]))

    # create a Bokeh plot
    plot = figure(title='Interactive Scatter Plot', tools='pan,box_zoom,reset')
    plot.circle(x='x', y='y', size='size', source=source, color='blue')

    # create a slider widget
    slider = Slider(start=10, end=50, step=1, value=20, title='Circle Size')

    # define a callback function for the slider
    def update_size(attrname, old, new):
        source.data['size'] = [new * 3, new * 6, new * 9]

    # add the slider widget to the layout and define its callback
    slider.on_change('value', update_size)
    layout = column(plot, slider)

    # set the app layout
    curdoc().add_root(layout)
```

```python
Bokeh provides several ways to embed Bokeh plots into web applications built with Flask or Django. 

Here are two approaches:-

# In Flask

#1
Using bokeh.embed.components:
    
from flask import Flask, render_template
from bokeh.plotting import figure, output_file, show
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    plot = figure(plot_width=400, plot_height=400)
    plot.circle([1, 2, 3], [4, 5, 6], size=10)
    script, div = components(plot)
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run()

    
#2
Using bokeh.embed.json_item

from flask import Flask, render_template
from bokeh.plotting import figure, output_file, show
from bokeh.embed import json_item
import json

app = Flask(__name__)

@app.route('/')
def index():
    plot = figure(plot_width=400, plot_height=400)
    plot.circle([1, 2, 3], [4, 5, 6], size=10)
    plot_json = json.dumps(json_item(plot))
    return render_template('index.html', plot_json=plot_json)

if __name__ == '__main__':
    app.run()
```
