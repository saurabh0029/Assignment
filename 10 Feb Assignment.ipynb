```python
Q1. Ans


Open() function is used to open a file.

Different modes of opening a file are:-


i)Read Only (‘r’): Open text file for reading. The handle is positioned at the beginning of the file.
If the file does not exist, raises an I/O error. This is also the default mode in which the file is opened.

ii)Read and Write (‘r+’): Open the file for reading and writing. The handle is positioned at the beginning of the file.
Raises I/O error if the file does not exist.

iii)Write Only (‘w’): Open the file for writing. For the existing files, the data is truncated and over-written.
The handle is positioned at the beginning of the file. Creates the file if the file does not exist.

iv)Write and Read (‘w+’): Open the file for reading and writing. For existing files, data is truncated and over-written.
The handle is positioned at the beginning of the file.

v)Append Only (‘a’): Open the file for writing. The file is created if it does not exist.The handle is positioned at the end
of the file. The data being written will be inserted at the end, after the existing data.

vi)Append and Read (‘a+’): Open the file for reading and writing. The file is created if it does not exist. The handle is
positioned at the end of the file. The data being written will be inserted at the end, after the existing data.

vii)Read Only in Binary format(‘rb’):  It lets the user open the file for reading in binary format.

viii)Read and Write in Binary Format(‘rb+’): It lets the user open the file for reading and writing in binary format.

ix)Write Only in Binary Format(‘wb’): It lets the user open the file for writing in binary format. When a file gets opened
in this mode, there are two things that can happen mostly. A new file gets created if the file does not exist. 
The content within the file will get overwritten if the file exists and has some data stored in it. 

x)Write and Read in Binary Format(‘wb+’):  It lets the user open the file for reading as well as writing in binary format.
When a file gets opened in this mode, there are two things that can mostly happen. A new file gets created for writing and
reading if the file does not exist. The content within the file will get overwritten if the file exists and has some data
stored in it.

xi)Append only in Binary Format(‘ab’): It lets the user open the file for appending in binary format. A new file gets created
if there is no file. The data will be inserted at the end if the file exists and has some data stored in it. 

xii)Append and Read in Binary Format(‘ab+’): It lets the user open the file for appending and reading in binary format.
A new file will be created for reading and appending if the file does not exist. We can read and append if the file 
exists and has some data stored in it.  

```

```python
Q2. Ans

Actually, Close() function is used to close a open file.

Because files are limited resources managed by the operating system, making sure files are closed after use will protect
against hard-to-debug issues like running out of file handles or experiencing corrupted data.
```

```python
Q3. Ans

f=open("Goal.txt","w")                # Open a File / Create

f.write("I want to become a Data Scientist.")  # Write

f.close()                          # Close a File


with open("Goal.txt","r") as f:       # Open in a read mode
    
    read_file=f.read()        # Read the Data
    
print(read_file)                  # Printing the Data

```

```python
Q4. Ans

read() :- The read() method returns the specified number of bytes from the file. Default is -1 which means the whole file.

readline() :- The readline() method returns one line from the file. We can also specified how many bytes from the line to
return,by using the size parameter.

readlines() :- The readlines() method returns a list containing each line in the file as a list item. Use the hint parameter to
limit the number of lines returned. If the total number of bytes returned exceeds the specified number, no more lines are 
returned.




# Explain using code

with open("Introduction.txt","w") as file:       # Open  file in write mode
    
    file.write("My Name is Saurabh Kumar. \nI am from  Siwan , Bihar.")
    

    
with open("Introduction.txt","r") as file:       # Open file in read mode
    
    intro=file.read()
    
print("Using read() method :- ",intro)    # Output will be Using read() method :-  My Name is Saurabh Kumar. 
                                                                                 # I am from  Siwan , Bihar.
    
    
with open("Introduction.txt","r") as file:
    
    intro1=file.readline()                
    
print("\n Using readline() method :-",intro1)   # Output will be Using readline() method :- My Name is Saurabh Kumar.

with open("Introduction.txt","r") as file:
    
    intro2=file.readlines()                
    
print("Using readlines() method :-",intro2)   # Output will be Using readlines() method :- ['My Name is Saurabh Kumar. \n', 'I am from  Siwan , Bihar.']


```

```python
Q6. Ans

# write()

The write() method writes a specified text to the file.

Where the specified text will be inserted depends on the file mode and stream position.

"a":  The text will be inserted at the current file stream position, default at the end of the file.

"w": The file will be emptied before the text will be inserted at the current file stream position, default 0.


For Example

with open("Doc.txt","w") as d:
    
    d.write("Hello, Sudhanshu Sir")
    
with open("Doc.txt","a") as d:
    
    d.write(" What's going on ?")
    
    
# writelines()

The writelines() method writes the items of a list to the file.

Where the texts will be inserted depends on the file mode and stream position.

"a":  The texts will be inserted at the current file stream position, default at the end of the file.

"w": The file will be emptied before the texts will be inserted at the current file stream position, default 0.


For Example

with open("Self_intro.txt","w") as d:
    
    d.writelines(["My name is Saurabh","\n I live in Siwan ,Bihar"])
    

with open("Self_intro.txt","a") as d:
    
    d.writelines(["\n My hobby is playing games","\n I aspire to be an entrepreneur" ])
    
    
```
