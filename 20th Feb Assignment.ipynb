```python
Q4. Ans 

# What are templates in Flask?

Templates are files that contain static data as well as placeholders for dynamic data. In Flask, templates are used to 
generate HTML pages dynamically. A template is essentially an HTML file with placeholders for dynamic content. When a 
Flask application serves a webpage, it uses a template to generate the HTML for that page. The dynamic content can be passed 
to the template as variables, and the placeholders are replaced with the actual content when the template is rendered.

# Why is the render_template() function used?

The render_template() function is  used to render the external HTML file to be returned as the response from the view function 
or response to a client's request.
```
