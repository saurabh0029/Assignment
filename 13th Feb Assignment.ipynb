```python
Q3.Ans


ArithmeticError is simply an error that occurs during numeric calculations.

ArithmeticError types in Python include:

OverFlowError
ZeroDivisionError
FloatingPointError


# ZeroDivision Error

import logging

logging.basicConfig(filename="errors.log",level=logging.ERROR)

try:
    
    a=4/0
    
except ArithmeticError as e:   # we can also use ZeroDivisionError
    
    logging.error("I am trying to handle ZeroDivision Error   {}".format(e))
    
    
    
    
# OverFlowError

import logging

logging.basicConfig(filename="New_Error.log",level=logging.ERROR)


j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    logging.error("I am try to handle Arithmetic Error {} {}".format((e),(e.__class__)))
    
```

```python
Q4. Ans


Because, Lookup Error acts as a base class for the exceptions that occur when a key or index used on a mapping or sequence
of a list/dictionary is invalid or does not exists.

The two types of exceptions raised are:

KeyError
IndexError

# KeyError

If a key we are trying to access is not found in the dictionary, a key error exception is raised.

for example:-

import logging



logging.basicConfig(filename="Output.log", level=logging.INFO)

try:
    dic={1:"Rahul",2:"Rohan"}
   
    logging.info("I am trying to print data %s".format(dic[4]))
                                                       
                                                     
                                                       
except LookupError as e:
                                                       
    logging.error("Key not found {}".format((e)))
    
    
    
# IndexError

When we are trying to access an index (sequence) of a list that does not exist in that list or is out of range
of that list, an index error is raised.

For example:-

import logging

logging.basicConfig(filename="Output.log",level=logging.INFO)

try:  
    a = ['a', 'b', 'c']  
    
    logging.info("I am trying to print a[4]")
    
    print (a[4])  
    
except LookupError:  
    logging.error("Index Error Exception Raised, list index out of range")
    
```
