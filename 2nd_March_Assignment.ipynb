```python
#What is Matplotlib?

Matplotlib is a popular data visualization library orthe basic plotting library  for Python. It provides a wide
range of tools for creating  high-quality graphs, charts, andother visualizations. Matplotlib is used extensively
in data science, scientific research, and other fields where data analysis and visualizationare important.

#Why is it used?

It is used for creating graphs , charts , and many other visualization.

#Name five plots that can be plotted using the Pyplot module of Matplotlib.

(i) Line plot: A line plot is used to visualize data as a series of points connected by a line. It is commonly used
               to show how a variable changes over time.

(ii) Bar plot: A bar plot is used to compare different categories of data. It consists of rectangular bars, where
               the height of each bar  corresponds to the value of the data being plotted.

(iii) Scatter plot: A scatter plot is used to show the relationship between two variables. It consists of points
                    that are plotted based on  their x and y coordinates.

(iv) Histogram: A histogram is used to show the distribution of a dataset. It consists of bars that represent the
                frequency of data falling within certain intervals.

(v) Pie chart: A pie chart is used to show the proportion of each category in a dataset. It consists of slices of
               a circle, where the size of each slice corresponds to the proportion of data in that category.
    
```

```python
#What is a scatter plot?

A scatter plot is a type of plot used to visualize the relationship between two variables. It is created by plotting
individual data points on  a two-dimensional graph, where one variable is plotted on the x-axis and the other 
variable is plotted on the y-axis.

import numpy as np
import matplotlib.pyplot as mtpl

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))


mtpl.scatter(x,y)  # creating Scatter for the given data

mtpl.xlabel('X-axis')
mtpl.ylabel('Y-axis')
mtpl.title('Scatter Plot for Ramdon Data')
```

```python
# Why sub plot function is used?

The subplot() function is used in Matplotlib, a Python plotting library, to create multiple plots in a single figure.
It allows you to divide the figure into a grid of rows and columns and specify which grid cell a particular plot 
should occupy.

```

```python
import numpy as np
import matplotlib.pyplot as mtpl
#For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

x= np.array([0, 1, 2, 3, 4, 5])
y1= np.array([0, 100, 200, 300, 400, 500])

#For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

y2= np.array([50, 20, 40, 20, 60, 70])

#For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

y3= np.array([10, 20, 30, 40, 50, 60])

#For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

y4= np.array([200, 350, 250, 550, 450, 150])



#Create the  line 1 subplot

mtpl.subplot(2, 2, 1)
mtpl.plot(x, y1)
mtpl.title('Line 1 Plot')

#Create the line 2 subplot

mtpl.subplot(2, 2, 2)
mtpl.plot(x, y2)
mtpl.title('Line 2 Plot')


#Create the  line 3 subplot

mtpl.subplot(2, 2, 3)
mtpl.plot(x, y3)
mtpl.title('Line 3 Plot')

#Create the  line 4 subplot


mtpl.subplot(2, 2, 4)
mtpl.plot(x, y4)
mtpl.title('Line 4 Plot')

#show the plots

mtpl.show()
```

```python
# What is Bar Plot?

A bar plot, also known as a bar chart, is a type of graph that represents categorical data with rectangular bars. 
Each bar represents a category,  and the height or length of the bar corresponds to the value or frequency of the
category.

# Why is it used?

Bar plots are often used to visualize and compare the numerical values of different categories or groups. They are
useful for displaying data that can be divided into discrete categories, such as the number of people in different
age groups, the frequency of different outcomes in a survey, or the sales performance of different products.
```

```python
import numpy as np
import matplotlib.pyplot as mtpl
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])


mtpl.bar(company,profit)  #Crerate bar plot
mtpl.xlabel("Company")
mtpl.ylabel("Profit")

mtpl.show()
```
